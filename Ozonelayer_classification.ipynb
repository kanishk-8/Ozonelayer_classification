# **Ozone layer classification**

## **Importing dataset and libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import RidgeClassifier
from sklearn.naive_bayes import GaussianNB
```

### **Reading csv file and cleaning data**

```python
data = pd.read_csv('/content/eighthr.data',header=None)
```

```python
data.head()
```

```python
data.drop([0],axis=1,inplace=True)
```

```python
data
```

**Replacing "?" with nan**

```python
for i in data.columns:
    data[i] = data[i].replace('?',np.nan)
```

```python
data.info()
```

```python
data.describe()
```

**Replacing missing values with mean**

```python
from sklearn.impute import SimpleImputer
```

```python
for i in data.columns[:-1]:
  data[i] = data[i].astype(float)
```

```python
imputer = SimpleImputer(strategy='mean')
data_imputed = pd.DataFrame(imputer.fit_transform(data), columns=data.columns)
```

```python
data_imputed
```

**Creating x and y variables to use in the model**

```python
x = data_imputed.drop([73], axis=1)
y = data_imputed[73]
```

### **Spliting data into train and test**

```python
xtrain, xtest, ytrain, ytest= train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
xtrain
```

## Random forest classifier

**Training model**

```python
rf = RandomForestClassifier(max_depth=2, random_state=0)
rf.fit(xtrain, ytrain )
```

**Prediction on Test Data**

```python
ypredict_rfc = rf.predict(xtest)
```

**Accuracy Score of the model**

```python
accuracy_score(ytest, ypredict_rfc)
```

## K nearest neighbors

**Training the model**

```python
reg= KNeighborsClassifier(n_neighbors=6)
reg.fit(xtrain, ytrain)
```

**Prediction on test data**

```python
ypredict_knn = reg.predict(xtest)
```

**Accuracy score of model**

```python
accuracy_score(ytest, ypredict_knn)
```

## Descision Tree Classifier

**Training the model**

```python
tree = DecisionTreeClassifier(random_state=0)
tree.fit(xtrain, ytrain)
```

**Prediction on Test Data**

```python
ypredict_dtc = tree.predict(xtest)
```

**Accuracy of the model**

```python
accuracy_score(ytest, ypredict_dtc)
```

## Ridge classifier

**Training the model**

```python
clf = RidgeClassifier()
clf.fit(xtrain, ytrain)
```

**Prediction on test data**

```python
ypredict_ridge = clf.predict(xtest)
```

**Accuracy of the model**

```python
accuracy_score(ytest, ypredict_ridge)
```

## Naive bayes classifier

**Training the model**

```python
gnb = GaussianNB()
gnb.fit(xtrain, ytrain)
```

**Prediction on test data**

```python
ypredict_gnb = gnb.predict(xtest)
```

**Accuracy of the model**

```python
accuracy_score(ytest, ypredict_gnb)
```
